```python
# Load the data
df1 = pd.read_csv(Path('LoanStats_2019Q1.csv.zip'), skiprows=1)[:-2]
df2 = pd.read_csv(Path('LoanStats_2019Q2.csv.zip'), skiprows=1)[:-2]
df3 = pd.read_csv(Path('LoanStats_2019Q3.csv.zip'), skiprows=1)[:-2]
df4 = pd.read_csv(Path('LoanStats_2019Q4.csv.zip'), skiprows=1)[:-2]

df = pd.concat([df1, df2, df3, df4]).loc[:, columns].copy()
```

```python
# convert interest rate to numerical
df['int_rate_new'] = df['int_rate'].str.replace('%', '')
df['int_rate_new'] = df['int_rate_new'].astype('float') / 100
df[['int_rate', 'int_rate_new']]
```

```python
x = {'Current': 'low_risk'}   
df = df.replace(x)
df
```

```python
df['loan_status'].unique()
```

```python

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()

# Remove the `Issued` loan status
issued_mask = df['loan_status'] != 'Issued'
df = df.loc[issued_mask]

# convert interest rate to numerical
df['int_rate'] = df['int_rate'].str.replace('%', '')
df['int_rate'] = df['int_rate'].astype('float') / 100


# Convert the target column('loan_status') values to low_risk and high_risk based on their values

# the following is same as df.loc[df['loan_status'] == 'Current', 'loan_status'] = 'low_risk'
# looks for 'Current' in any field and replaces with 'low_risk'
x = {'Current': 'low_risk'}   
df = df.replace(x)

#  Convert the target column('loan_status') values to low_risk high_risk for the following
x = dict.fromkeys(['Late (31-120 days)', 'Late (16-30 days)', 'Default', 'In Grace Period'], 'high_risk')    
df = df.replace(x)
```

```python
df.columns
```

## Create seperate dataframes for low risk and high risk loans

```python

low_risk_rows = df[df[target] == 'low_risk']
len(low_risk_rows)
```

```python
high_risk_rows = df[df[target] == 'high_risk']
len(high_risk_rows)
```

```python
# the dataset has more low_risk records than high_risks 
# so we are going to make them even by picking equal number of records randomly
```

```python
# select equal number of randomly selected records from low_risk_rows as the number of records in len(high_risk_rows)

sample_low_risk_df = low_risk_rows.sample(n=len(high_risk_rows), random_state=42)
sample_low_risk_df
```

```python
# Join selected low_risk_rows(sample_low_risk_df) and high_risk_rows to a single dataframe
#df = pd.concat([low_risk_rows, high_risk_rows.sample(n=len(low_risk_rows), replace=True)])
df = pd.concat([sample_low_risk_df, high_risk_rows])
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
# change the field name 'loan_status' to 'target'
df = df.rename({target:'target'}, axis="columns")
df
```

## write cleaned and undersampled data to a csv
